The types of DoS attacks that we focus on simulating include: 

1. SYN Flood Attack: This attack overwhelms the target system by sending a flood 

of TCP SYN packets. 

2. Ping of Death (POD) Attack: The POD attack sends oversized ICMP packets to 

the target, with the intent of crashing or disrupting the target system. 

3. UDP Flood Attack: This attack overwhelms the target by sending a very high rate of UDP 

 Datagrams often to random or specific ports. The target must repeatedly check for a listening 

 application and, if none exists, generate ICMP “Destination Unreachable” replies, consuming 

 CPU and bandwidth and potentially resulting in denial of service. 

 

The main objectives of this project are: 

1. Simulate DoS attacks using Python scripts and Scapy. 

2. Detect the attacks in real-time through packet sniffing techniques. 

3. Log the detected attacks into a CSV file for further analysis. 

4. Analyze attack traffic to derive insights such as identifying the most damaging 

     attack, geographical location of attackers, and traffic behavior over time.

# DoS/DDoS Analysis from `ddos_attacks.csv`
This notebook loads **`ddos_attacks.csv`**
enriches public source IPs using **ip-api.com** to get countries, and answers:

1. **Where are the suspect IP addresses located?**
2. **Which attack generated more packets?** (bar chart)
3. **Which attack generated the largest traffic** in **bytes** and in **bytes/sec** (bar charts)

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

CSV = Path('ddos_attacks.csv')
assert CSV.exists(), 'Put ddos_attacks.csv in the same folder as this notebook.'
df = pd.read_csv(CSV)
print('Loaded rows:', len(df))

orig_cols = list(df.columns)
df.columns = [c.strip().lower().replace(' ', '_') for c in df.columns]

rename_map = {
    'tsstart':'ts_start', 'ts_start':'ts_start', 'start':'ts_start',
    'tsend':'ts_end', 'ts_end':'ts_end', 'end':'ts_end',
    'srcip':'src_ip', 'source_ip':'src_ip', 'src':'src_ip',
    'dstip':'dst_ip', 'destination_ip':'dst_ip', 'dst':'dst_ip',
    'attacktype':'attack_type', 'type':'attack_type', 'label':'attack_type',
    'pktcount':'pkt_count', 'packets':'pkt_count', 'packet_count':'pkt_count',
    'bytecount':'byte_count', 'bytes':'byte_count', 'total_bytes':'byte_count',
    'ratebps':'rate_bps', 'bps':'rate_bps', 'bytes_per_second':'rate_bps',
    'ratepps':'rate_pps', 'pps':'rate_pps'
}
df = df.rename(columns={k:v for k,v in rename_map.items() if k in df.columns})

# Coerce numerics when present
for col in ['pkt_count','byte_count','rate_bps','rate_pps']:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

if 'rate_bps' not in df.columns:
    try:
        if {'ts_start','ts_end','byte_count'}.issubset(df.columns):
            t0 = pd.to_datetime(df['ts_start'], errors='coerce')
            t1 = pd.to_datetime(df['ts_end'], errors='coerce')
            secs = (t1 - t0).dt.total_seconds().replace(0, np.nan)
            df['rate_bps'] = (df['byte_count'] * 8 / secs).fillna(0)
            print('Computed rate_bps from byte_count and window duration.')
    except Exception as e:
        print('Could not compute rate_bps:', e)

df.head(20)
```

## Country enrichment via ip-api.com (batch with cache)

```python
import requests, time, json
from ipaddress import ip_address

def is_public_ipv4(ip):
    try:
        ipobj = ip_address(str(ip))
        return (ipobj.version==4) and (not ipobj.is_private) and (not ipobj.is_loopback) and (not ipobj.is_link_local) and (not ipobj.is_reserved)
    except Exception:
        return False

unique_public = sorted({str(ip) for ip in df.get('src_ip', pd.Series([],dtype=str)).astype(str) if is_public_ipv4(ip)})
print('Unique public src IPs:', len(unique_public))

cache_path = Path('data/ipapi_cache.json'); cache_path.parent.mkdir(parents=True, exist_ok=True)
cache = {}
if cache_path.exists():
    try:
        cache = json.loads(cache_path.read_text())
    except Exception:
        cache = {}

to_lookup = [ip for ip in unique_public if ip not in cache]
print('Need to lookup:', len(to_lookup))

def ipapi_batch_lookup(ips):
    url = 'http://ip-api.com/batch?fields=status,country,query'
    r = requests.post(url, json=[{'query': ip} for ip in ips], timeout=10)
    r.raise_for_status()
    return r.json()

BATCH = 100
for i in range(0, len(to_lookup), BATCH):
    chunk = to_lookup[i:i+BATCH]
    try:
        data = ipapi_batch_lookup(chunk)
        for row in data:
            q = row.get('query')
            if row.get('status') == 'success':
                cache[q] = {'country': row.get('country','Unknown')}
            else:
                cache[q] = {'country': 'Unknown'}
        cache_path.write_text(json.dumps(cache, indent=2))
    except Exception as e:
        print('Lookup failed for chunk, continuing:', e)
    time.sleep(1)

def country_or_bucket(ip):
    s = str(ip)
    if s in cache:
        return cache[s].get('country','Unknown')
    try:
        ipobj = ip_address(s)
        if ipobj.is_loopback:  return 'Loopback'
        if ipobj.is_private:   return 'Private (LAN/VPN)'
        if ipobj.is_link_local or ipobj.is_reserved: return 'Link-local/Reserved'
        return 'Unknown'
    except Exception:
        return 'Unknown'

df['country'] = df['src_ip'].astype(str).apply(country_or_bucket)
out_geo = Path('data/attacks_geo.csv'); out_geo.parent.mkdir(parents=True, exist_ok=True)
df.to_csv(out_geo, index=False)
print('Enriched CSV saved to', out_geo.resolve())
df[['src_ip','country']].head()
```

## Q1) Where are the suspect IP addresses located? (Text + Plot)

```python
by_country = df.groupby('country', dropna=False)['src_ip'].nunique().sort_values(ascending=False)
print('Text answer — top countries/categories by unique source IPs:')
display(by_country.head(10))

ax = by_country.head(20).plot(kind='bar', figsize=(10,4), title='Unique suspect IPs by country (or category)')
ax.set_xlabel('Country / Category'); ax.set_ylabel('Unique source IPs')
plt.tight_layout(); plt.show()
```

Map plot 

```python
# pip install plotly
import plotly.express as px

ignore = {"Private (LAN/VPN)", "Loopback", "Link-local/Reserved", "Unknown", "Unknown (no geo)"}
counts = (
    df.loc[~df["country"].isin(ignore)]
      .groupby("country")["src_ip"]
      .nunique()
      .reset_index(name="count")
)

if counts.empty:
    print("No public IP countries to plot. Run the ip-api enrichment cell and ensure there are public src_ip values.")
else:
    fig = px.choropleth(
        counts,
        locations="country",
        locationmode="country names",
        color="count",
        title="Suspect IPs by Country (public IPs only)"
    )
    fig.show()
```

## Q2) Which attack generated more packets? (Text + Plot)

```python
assert 'attack_type' in df.columns, 'Need an attack_type column in ddos_attacks.csv'
assert 'pkt_count' in df.columns, 'Need a pkt_count (packets per window) column in ddos_attacks.csv'
by_attack_packets = df.groupby('attack_type')['pkt_count'].sum().sort_values(ascending=False)
print('Text answer — most packets by attack type:')
display(by_attack_packets)
ax = by_attack_packets.plot(kind='bar', figsize=(6,4), title='Total packets by attack type')
ax.set_xlabel('Attack type'); ax.set_ylabel('Total packets')
plt.tight_layout(); plt.show()
```

## Q3) Which attack generated the largest traffic (bytes) and (bytes/sec)? (Text + Plots)

```python
assert 'byte_count' in df.columns, 'Need a byte_count column in ddos_attacks.csv'
by_attack_bytes = df.groupby('attack_type')['byte_count'].sum().sort_values(ascending=False)
print('Text answer — largest total bytes by attack type:')
display(by_attack_bytes)
ax = by_attack_bytes.plot(kind='bar', figsize=(6,4), title='Total bytes by attack type')
ax.set_xlabel('Attack type'); ax.set_ylabel('Total bytes')
plt.tight_layout(); plt.show()

if 'rate_bps' not in df.columns:
    print('Note: rate_bps missing; computed earlier if timestamps present. If still missing, add rate_bps to CSV to answer bytes/sec precisely.')
else:
    by_attack_bps_mean = df.groupby('attack_type')['rate_bps'].mean().sort_values(ascending=False)
    by_attack_bps_max  = df.groupby('attack_type')['rate_bps'].max().sort_values(ascending=False)
    print('Text answer — highest MEAN bytes/sec by attack type:')
    display(by_attack_bps_mean)
    print('Text answer — highest PEAK bytes/sec by attack type:')
    display(by_attack_bps_max)
    ax = by_attack_bps_mean.plot(kind='bar', figsize=(6,4), title='Mean bytes/sec by attack type')
    ax.set_xlabel('Attack type'); ax.set_ylabel('Mean bytes/sec')
    plt.tight_layout(); plt.show()
```
